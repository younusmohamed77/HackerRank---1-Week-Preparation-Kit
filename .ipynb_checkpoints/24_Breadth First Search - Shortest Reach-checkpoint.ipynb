```python
from collections import deque, defaultdict

def bfs(n, m, edges, s):
    # Initialize the graph as adjacency list
    graph = defaultdict(list)
    
    # Build the graph
    for u, v in edges:
        graph[u].append(v)
        graph[v].append(u)
    
    # Initialize distances array with -1 indicating unreachable
    distances = [-1] * (n + 1)
    distances[s] = 0
    
    # BFS initialization
    queue = deque([s])
    
    while queue:
        node = queue.popleft()
        current_distance = distances[node]
        
        for neighbor in graph[node]:
            if distances[neighbor] == -1:  # If neighbor is not visited
                distances[neighbor] = current_distance + 6
                queue.append(neighbor)
    
    # Convert distances to match the required output format
    # Skip the dummy element at index 0 and replace 0 with -1
    return [distances[i] if i != s else -1 for i in range(1, n + 1)]

if __name__ == '__main__':
    # Read number of queries
    q = int(input().strip())
    
    results = []
    
    for _ in range(q):
        # Read each query
        first_multiple_input = input().strip().split()
        n = int(first_multiple_input[0])  # number of nodes
        m = int(first_multiple_input[1])  # number of edges
        
        edges = []
        for _ in range(m):
            edges.append(list(map(int, input().strip().split())))  # read each edge
        
        s = int(input().strip())  # starting node
        
        # Perform BFS to get shortest distances
        result = bfs(n, m, edges, s)
        
        # Store result for current query
        results.append(result)
    
    # Print all results as expected output format
    for result in results:
        print(" ".join(map(str, result)))
```

```python
from collections import deque, defaultdict

def bfs(n, m, edges, s):
    # Initialize the graph as adjacency list
    graph = defaultdict(list)
    
    # Build the graph
    for u, v in edges:
        graph[u].append(v)
        graph[v].append(u)
    
    # Initialize distances array with -1 indicating unreachable
    distances = [-1] * (n + 1)
    distances[s] = 0
    
    # BFS initialization
    queue = deque([s])
    
    while queue:
        node = queue.popleft()
        current_distance = distances[node]
        
        for neighbor in graph[node]:
            if distances[neighbor] == -1:  # If neighbor is not visited
                distances[neighbor] = current_distance + 6
                queue.append(neighbor)
    
    # Prepare the result list, excluding the start node
    result = []
    for i in range(1, n + 1):
        if i != s:
            result.append(distances[i])
    
    return result
```
